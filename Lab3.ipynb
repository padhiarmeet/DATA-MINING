### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
val = ['PassengerId','Survived','Pclass','Age','SibSp','Parch']
# print('PassangerId:')
# print('\t Mean = ', data.PassengerId.mean())
# print('\t Mean = ', data.PassengerId.std())
# print('\t Mean = ', data.PassengerId.min())
# print('\t Mean = ', data.PassengerId.max())

for i in val :
    print(i,':')
    print('\t Mean = ', data[i].mean())
    print('\t Mode = ', data[i].mode()[0])
    print('\t Standard Deviation = ', data[i].std())
    print('\t Min = ', data[i].min())
    print('\t Max = ', data[i].max())
    print('\t Range = ', data[i].max() - data[i].min())
```
