### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
pure_data = data.copy()
```

```python
pure_data['Age'] = pure_data['Age'].interpolate()
```

```python
# Data Normalization, Feature Scaling, Outlier DItection
pure_data['New_Age'] = ((pure_data['Age'] - pure_data['Age'].min())/(pure_data['Age'].max() - pure_data['Age'].min()))
```
